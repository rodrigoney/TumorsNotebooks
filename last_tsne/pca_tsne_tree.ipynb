```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import pylab as Plot
import matplotlib.cm as cm
import sys
```

```python
raw_df = pd.read_csv("input/Planilha_RODRIGO_tumores.csv")

X = raw_df[['CD45','CD56','CD3','CD19']].values
labels = raw_df.hematopoietico.values

last_divergence = sys.maxint
X_embedded = None
for i in range(0,25):
    my_tse = TSNE(n_components=2)
    cur_X_embedded = my_tse.fit_transform(X)
    if last_divergence >= my_tse.kl_divergence_:
        X_embedded = cur_X_embedded
        last_divergence = my_tse.kl_divergence_
        
colors_map = []
for hematopoetico in labels:
    if hematopoetico == 'sim':
        colors_map.append('blue')
    else:
        colors_map.append('red')
        
print 'Lowest kl_divergence: ' + str(last_divergence)
Plot.scatter(cur_X_embedded[:,0], cur_X_embedded[:,1], label = labels, color=colors_map);
Plot.show();
```

```python
labels = raw_df.diagnostico.values
unique_labels = raw_df.diagnostico.unique()
colors_map = []
all_cmap = plt.cm.get_cmap('hsv', len(unique_labels))
for disease in labels:
    for index in range(0, len(unique_labels)):
        if disease == unique_labels[index]:
            colors_map.append(all_cmap(index))
            break
        
        
print 'Lowest kl_divergence: ' + str(last_divergence)
used_labels = {}
for i in range(0, len(cur_X_embedded)):
    if labels[i] not in used_labels:
        plt.scatter(cur_X_embedded[i][0], cur_X_embedded[i][1], label = labels[i], color=colors_map[i])
        used_labels[labels[i]] = True
    else:
        plt.scatter(cur_X_embedded[i][0], cur_X_embedded[i][1], color=colors_map[i])


plt.rcParams['figure.figsize'] = 15, 16
plt.title('TSNE Result')
plt.legend(scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=9)
plt.show()
```

```python
from sklearn.decomposition import PCA
raw_df = pd.read_csv("input/Planilha_RODRIGO_tumores.csv")
X = raw_df[['CD45','CD56','CD3','CD19']].values
labels = raw_df.hematopoietico.values

pca = PCA(n_components=2)
pca.fit(X)

print 'PCA - Number of components: ' + str(pca.n_components_)
print ''
print 'PCA - Components Values: ' + str(pca.components_)
print ''
print 'PCA - Explained Variance (The amount of variance explained by each of the selected components. Equal to n_components largest eigenvalues of the covariance matrix of X): ' + str(pca.explained_variance_)
print ''
print 'PCA - Means (Per-feature empirical mean, estimated from the training set.): ' + str(pca.mean_)
print ''
print 'PCA - Noise Variance (The estimated noise covariance following the Probabilistic PCA model from Tipping and Bishop 1999): ' + str(pca.noise_variance_)
print ''
print 'PCA - Explained Variance Ratio (Percentage of variance explained by each of the selected components): ' + str(pca.explained_variance_ratio_)
print ''
print 'PCA - Singular Values (The singular values corresponding to each of the selected components. The singular values are equal to the 2-norms of the n_components variables in the lower-dimensional space.): ' + str(pca.singular_values_)

cur_X_embedded = pca.transform(X)

colors_map = []
for hematopoetico in labels:
    if hematopoetico == 'sim':
        colors_map.append('blue')
    else:
        colors_map.append('red')

Plot.scatter(cur_X_embedded[:,0], cur_X_embedded[:,1], label = labels, color=colors_map);
Plot.show();
```
